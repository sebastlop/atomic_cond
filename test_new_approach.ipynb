```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
x = np.logspace(1,4, 20)
y = np.log(x-7)/x
z_ion = np.arange(1,10)
dataset = pd.DataFrame()
dataset['Z_ion'] = np.array([zp*np.ones_like(x) for zp in np.arange(1,10)]).reshape(-1).astype(int)
dataset['LogE'] = np.log10(np.tile(x,9))
dataset['Logcs'] = np.log10(np.array([zp**2*np.log(x)/x*1e-15 for zp in range(1,10)]).reshape(-1))
dataset.head()
```

```python
for i in range(0,9):
    plt.plot(dataset[dataset['Z_ion']==i]['LogE'],dataset[dataset['Z_ion']==i]['Logcs'], '*')
```
