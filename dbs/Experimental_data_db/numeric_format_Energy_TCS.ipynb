```python
import pandas as pd
import numpy as np

# Input and output filenames
input_csv = "exp_TCS_database_single_ionization_atoms_raw.csv"
output_csv = "exp_TCS_database_single_ionization_atoms.csv"

# Load CSV
df = pd.read_csv(input_csv)

# Function to format numbers in scientific notation with 4 significant digits
def sci_not_4sig(x):
    try:
        return f"{x:.3e}"
    except:
        return x  # In case there are strings or NaNs

# Apply formatting
df["E_proj[keV/u]"] = df["E_proj[keV/u]"].apply(sci_not_4sig)
df["Cross_section[cm2]"] = df["Cross_section[cm2]"].apply(sci_not_4sig)

# Save to new CSV
df.to_csv(output_csv, index=False)

print(f"Reformatted file saved as {output_csv}")
```
